```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
```

```python
def continuous_peaks_ff(x, t):
    max0 = 0
    count = 0
    for x_i in x:
        if x_i == '0': 
            count += 1
        elif count > max0:
            max0 = count
            count = 0
    max1 = 0
    count = 0
    for x_i in x:
        if x_i == '1': 
            count += 1
        elif count > max1:
            max1 = count
            count = 0
    r = 0
    if max0 > t and max1 > t:
        r = len(x)
    return max(max0, max1) + r
```

```python
s = '1110000001'
continuous_peaks_ff(s, len(s)/10)
```

```python
x = [re.sub('0b', '', bin(x).zfill(11)) for x in range(500)]
y = [continuous_peaks_ff(x_1, 3) for x_1 in x]

plt.plot(range(500), y)
plt.show()
```
