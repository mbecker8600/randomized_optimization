```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
```

```python
def two_colors_ff(x):
    val = 0
    for i in range(1,len(x)-1):
        if x[i] != x[i-1] and x[i] != x[i+1]:
            val += 1
    return val
```

```python
x = [re.sub('0b', '', bin(x).zfill(11)) for x in range(500)]
y = [two_colors_ff(x_i) for x_i in x]

plt.plot(range(500), y)
plt.show()
```
